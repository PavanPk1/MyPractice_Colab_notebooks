```python
# Run this code cell
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

url = 'https://new-assets.ccbp.in/frontend/content/aiml/classical-ml/co2_emissions_canada.csv'

# Upload the data on colab and execute this code cell
data = pd.read_csv(url)
data.head()
```

**1. Find the top 3 vehicles with the highest CO2 emissions for each vehicle class.**

```python
# Write the code here
top_3_co2_emissions = data.groupby('Vehicle Class')['CO2 Emissions(g/km)'].nlargest(3)
print(top_3_co2_emissions)
```

**2. Find the top 3 car makes with the highest average CO2 emissions.**

```python
# Write the code here

top_3_makes = data.groupby('Make')['CO2 Emissions(g/km)'].mean().nlargest(3)
print(top_3_makes)
```

**3. Identify the fuel type that has the lowest average CO2 emissions.** In the given format:
- `Fuel type with lowest average CO2 emissions: N`

```python
# Write the code here

lowest_emission_fuel = data.groupby('Fuel Type')['CO2 Emissions(g/km)'].mean().idxmin()
print(f"Fuel type with lowest average CO2 emissions: {lowest_emission_fuel}")
```

**4. Add a new column for CO2 emissions per liter of fuel consumed (CO2 Emissions / Fuel Consumption) and list first 10 rows**

```python
data['CO2 per Liter'] = data['CO2 Emissions(g/km)'] / data['Fuel Consumption Comb (L/100 km)']
# display(data[['Make','Model','CO2 per Liter']].head(10))
print(data.loc[:, ['Make', 'Model', 'CO2 per Liter']].head(10))
```

**5. Find the car make with the highest fuel consumption on average.**

```python
# Write the code here

highest_fuel_make = data.groupby('Make')['Fuel Consumption Comb (L/100 km)'].mean().idxmax()
print(f"Car make with highest average fuel consumption: {highest_fuel_make}")
```

**6. Identify the top 3 car makes that have the most models with CO2 emissions above the 75th percentile.**

```python
# Write the code here

threshold = data['CO2 Emissions(g/km)'].quantile(0.75)
top_makes = data[data['CO2 Emissions(g/km)'] > threshold].groupby('Make').size().nlargest(3)
top_makes.name = 'Model'
print(top_makes)
```

**7. Create a new feature that categorizes vehicles into 'Low', 'Medium', and 'High' based on percentiles of the "CO2 Emissions(g/km)" column and get first 10 rows:**

- `Low: ≤ 33rd percentile`
- `Medium: 33rd to 66th percentile`
- `High: > 66th percentile`

```python
# Write the code here

data['Emission Category'] = pd.qcut(data['CO2 Emissions(g/km)'], q=[0, 0.33, 0.66, 1], labels=['Low', 'Medium', 'High'])
print(data[['Make','Model','CO2 Emissions(g/km)','Emission Category']].head(10))
```

**8. Determine if turbocharged engines (assuming turbocharged cars contain 'T' in model name) have higher average CO2 emissions.**

```python
turbo_vs_non_turbo = data.groupby(
    data['Model'].str.contains('T', case=False, na=False)
)['CO2 Emissions(g/km)'].mean()
turbo_vs_non_turbo.index.name = 'Turbocharged'
turbo_vs_non_turbo.name = 'CO2 Emissions(g/km)'
print(turbo_vs_non_turbo)
```

 **9. Compute the mean and standard deviation of CO2 emissions for each fuel type, and sort by highest variation.**

```python
import matplotlib.pyplot as plt

fuel_stats = data.groupby('Fuel Type')['CO2 Emissions(g/km)'].agg(['mean', 'std']).sort_values(by='std', ascending=False)
print(fuel_stats)

# Create the bar plot for mean CO2 emissions
plt.figure(figsize=(10, 6))
sns.barplot(x=fuel_stats.index, y='mean', data=fuel_stats, palette='viridis')

# Add error bars for standard deviation
plt.errorbar(x=fuel_stats.index, y=fuel_stats['mean'], yerr=fuel_stats['std'], fmt='none', c='black', capsize=5)

plt.title('Mean CO2 Emissions by Fuel Type with Standard Deviation')
plt.xlabel('Fuel Type')
plt.ylabel('CO2 Emissions (g/km)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**10. Categorize vehicles into engine size (5 bins), calculates the average CO2 emissions for each engine size bin and prints the engine size range that has the highest average CO2 emissions.**

```python
# Write the code here

# Categorize vehicles into 5 bins based on 'Engine Size(L)'
data['Engine Size Bin'] = pd.cut(data['Engine Size(L)'], bins=5)

# Calculate the average CO2 emissions for each engine size bin
engine_emission_bins = data.groupby('Engine Size Bin')['CO2 Emissions(g/km)'].mean()

# Find the engine size range that has the highest average CO2 emissions
highest_co2_bin = engine_emission_bins.idxmax()
highest_co2_value = engine_emission_bins.max()
print(f"Engine size range with highest CO2 emissions: {engine_emission_bins}")
```

**11. Compare the median fuel consumption between vehicles with engine sizes above and below the median engine size.** In the given format:
- `Above median engine size fuel consumption: ____`
- `Below median engine size fuel consumption: ____`

```python
# Write the code here
median_engine_size = data['Engine Size(L)'].median()
above_median_fuel = data[data['Engine Size(L)'] > median_engine_size]['Fuel Consumption Comb (L/100 km)']
below_median_fuel = data[data['Engine Size(L)'] <= median_engine_size]['Fuel Consumption Comb (L/100 km)']
print(f"Above median engine size fuel consumption: {above_median_fuel.median()}")
print(f"Below median engine size fuel consumption: {below_median_fuel.median()}")

# above = data['Engine Size(L)'].quantile(0.75)
# below = data['Engine Size(L)'].quantile(0.25)
# print(f"Above median engine size fuel consumption: {above}")
# print(f"Below median engine size fuel consumption: {below}")
```

**12. Find the vehicle with the best CO2 emission per horsepower ratio (assuming 'Engines Size' correlates with power).**

```python
# Write the code here
# Create a new column for CO2 emissions per unit of engine size (assuming it correlates with HP)
data['CO2 per HP'] = data['CO2 Emissions(g/km)'] / data['Engine Size(L)']

# Find the vehicle with the minimum CO2 per HP ratio
best_efficiency_vehicle = data.loc[data['CO2 per HP'].idxmin()]

# Display the Make, Model, and CO2 per HP for this vehicle
print(best_efficiency_vehicle.loc[['Make', 'Model', 'CO2 per HP']])

print("----------------------------------------")

# Find the index of the vehicle with minimum CO2 per HP ratio
min_index = data['CO2 per HP'].idxmin()

# Select the row as a DataFrame with relevant columns
best_efficiency_vehicle = data.loc[[min_index], ['Make', 'Model', 'CO2 per HP']]

# Display the DataFrame
print(best_efficiency_vehicle)
```

**13. Find the average CO2 emissions for each make, considering only models with fuel consumption above 10 L/100 km.**

```python
# Write the code here
high_fuel_cars = data[data['Fuel Consumption Comb (L/100 km)'] > 10]
avg_co2_per_make = high_fuel_cars.groupby('Make')['CO2 Emissions(g/km)'].mean()
print(avg_co2_per_make)
```
